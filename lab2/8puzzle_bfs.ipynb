```python
def bfs(initial,goal):
    queue=[]
    queue.append(initial)
    visited=[]

    while len(queue):
        state=queue.pop(0)
        visited.append(state)
        display(state)
        #print(state)
        
        if state==goal:
            print("Success!")
            return 
        
        neighbors=possible_moves(state,visited)
        
        for move in neighbors:
            if move not in queue and move not in visited:
                queue.append(move)
                
    print("Couldn't solve the puzzle :( ")        
```

```python
def possible_moves(state,visited): 
    #pos of empty slot
    b = state.index(0)
    m = []
    
    #Add all the possible directions
    if b not in [0,1,2]: 
        m.append('u')
    if b not in [6,7,8]: 
        m.append('d')
    if b not in [0,3,6]: 
        m.append('l')
    if b not in [2,5,8]: 
        m.append('r')
        
    possible_states = []
    
    for i in m:
        possible_states.append(gen(state,i,b))
        
    return [move for move in possible_states if move not in visited]
```

```python
def gen(state, m, b):
    temp = state.copy()                              
    
    if m=='d':
        temp[b+3],temp[b] = temp[b],temp[b+3]
    
    if m=='u':
        temp[b-3],temp[b] = temp[b],temp[b-3]
    
    if m=='l':
        temp[b-1],temp[b] = temp[b],temp[b-1]
    
    if m=='r':
        temp[b+1],temp[b] = temp[b],temp[b+1]
    
    return temp
```

```python
def display(state):
    for i in range (0,len(state),3):
        for j in range (i,i+3):
            if state[j]==0:
                print("_",end =" ")
            else:
                print(state[j],end =" ")
        print()
    print()
```

```python
#test1
initial=[1,2,5,3,4,0,6,7,8]
goal=[0,1,2,3,4,5,6,7,8]
bfs(initial,goal)
```
